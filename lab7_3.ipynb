In the file directory there is a file called `icecream.csv`. There are four columns.
The first column consists of the Github usernames of the people in the class. The next columns are empty and represent the persons favourite ice cream flavour, favourite colour and height in cm. The code below will read this file and produce graphs showing summaries of the class.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("icecream.csv")

# Plot icecream
plt.figure()
data["Favourite Ice Cream"].value_counts().plot(kind="pie")
plt.title("Favourite Ice Cream Flavours")
plt.xlabel("Ice Cream")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot colours
counts = data["Favourite Colour"].value_counts()
plt.figure()
counts.plot(kind="pie",colors=counts.index,autopct="%1.0f%%",)
plt.title("Favourite Colours")
plt.xlabel("Colour")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot heights
data["Height (cm)"] = pd.to_numeric(data["Height (cm)"], errors="coerce")
heights = data["Height (cm)"].dropna()

plt.figure()
plt.hist(heights, bins=10, edgecolor="black")
plt.title("Height Distribution")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
